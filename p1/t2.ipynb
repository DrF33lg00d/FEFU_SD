```python
import sqlite3
```

```python
def get_connect():
    connect = sqlite3.connect("bookshop.db")
    return connect

connect = get_connect()
```

```python
def get_cursor(connect):
    return connect.cursor()

cursor = get_cursor(connect)
```

```python
def select_author_table(cursor):
    sql = """SELECT name FROM author WHERE name = 'Ivan'
    """
    cursor.execute(sql)
    result = cursor.fetchall()
    return result
```

```python
for row in select_author_table(cursor):
    print(row[0])
```

```python
def fill_author_table(cursor):
    sql = """INSERT INTO author (name) VALUES
        ('Ivan'), ('Petr'), ('Victor')
    """
    cursor.execute(sql)
```

```python
def select_book_by_author(cursor, author_name):
    sql = """
        SELECT b.name, b.pages
        FROM book b
            INNER JOIN author a ON b.author_id = a.id
        WHERE 1=1
            and a.name = ?
    """
    cursor.execute(sql, (author_name,))
    result = cursor.fetchall()
    return result
```

```python
for row in select_book_by_author(cursor, 'Ivan Dev'):
    print(f'{row[0]} содержит {row[1]} страниц')
```

```python
def fill_book_table(cursor):
    """
    Функция заполняет таблицу book книгами и привязывает к авторам
    """
    sql = """INSERT INTO book (name, pages, author_id) VALUES
        ('Чистый код', 1523, 2),
        ('DevOps техники', 2314, 2),
        ('ПДД для чайников', 500, 3),
        ('Искусство', 413, 4)
    """
    cursor.execute(sql)
```

```python
fill_book_table(cursor)
connect.commit()
```

```python
def update_book_table(cursor, old_book_name, new_book_name, pages):
    sql = """UPDATE book
            SET name = ?, pages = ?
            WHERE 1=1
                AND name = ?
    """
    cursor.execute(sql, (new_book_name, pages, old_book_name))
```

```python
update_book_table(cursor, 'Чистый код. Издание 2', 'Чистый код. Издание 2', 13)
connect.commit()
```

```python
def delete_book_table(cursor, book_name):
    sql = """DELETE FROM book
            WHERE name = ?
    """
    cursor.execute(sql, (book_name,))
```

```python
delete_book_table(cursor, 'Чистый код. Издание 2')
connect.commit()
```
